```python
import os
from dotenv import load_dotenv

load_dotenv()  # .env 파일 로드

key = os.getenv("OPENAI_API_KEY")
print("키 존재:", key is not None)
print("길이:", len(key) if key else 0)
print(os.getenv("OPENAI_API_KEY"))
```

```python
import json
import openai

client = openai.OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

# n: 한 번의 요청으로 생성할 답변 개수 (여러 후보 중 선택 가능)
# temperature: 0~2, 높을수록 다양한 답변 (기본 1)
response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[{"role": "user", "content": "Hello, world!"}],
    n=5,  # 5개의 서로 다른 답변 생성
    temperature=2,  # 다양한 답변 유도
)

# 각 답변 출력 (enumerate start=1: 인덱스 1부터 시작)
for i, choice in enumerate(response.choices, start=1):
    print(f"--- 답변 {i} ---\n{choice.message.content}\n")

# 전체 응답을 JSON으로 확인하려면:
# print(json.dumps(response.model_dump(), indent=2, ensure_ascii=False))
```
